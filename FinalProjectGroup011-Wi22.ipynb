# Research Question

Is there a relationship between the biological features (height, weight) of the soccer players and their entrance to the English premier league and their performance?

We will analyze the trends of the height of all positions in the English premier league from the 1950s.

Also, we will analyze existing professional players' weight by positions and performance ratings measured by FIFA, international Football Association.

## Background & Prior Work

A similar study was conducted to find out the age of Olympic athletes that give the highest performance. [The 2016 research from Exercise Physiology Laboratory](https://www.researchgate.net/publication/312632883_Age_of_peak_performance_in_Olympic_sports_A_comparative_research_among_disciplines) measured the performance of 3548 athletes from various types of sports. The result of the study indicated that mostly younger athletes had higher performance than older athletes. Since the range of athletes’ age was 14.0 to 52.8 and the age of athletes who recorded high performance was mostly middle of the ’20s, the study proved that younger athletes have higher performance than older athletes.

[Another research conducted in 2019](https://pubmed.ncbi.nlm.nih.gov/31268996/) agrees with the study stated above. The study was to find out the effects of age on soccer players’ performance. They separated soccer players into 5 age groups: (<20 years, 20-24.9 years, 25-29.9 years, 30-34.9 years, and ≥35 years). The result of the study showed that older players’ performance is significantly less than the performance of younger players. For example, total distance, medium-speed running distance, high-speed running (HSR) distance, very HSR (VHSR) distance, sprint distance, and maximum running speed of older players were less than those of younger players. Therefore, the study indicated that younger soccer players have higher performance than older soccer players.

Although two researches show a strong relationship between the age of soccer players and their performance, they only consider a factor: age. It is true that researches stated above help answering our group’s question, but they do not directly answer our question. We also want to find out whether the height and weight of players affect players’ playing performance.

Here are some common functions to normalize our data into consistent format

```python
def extract_year(birth):
    if isinstance(birth, float) or isinstance(birth, int):
        try:
            return int(birth)       # 1950
        except:
            return np.nan
    if birth.count("/") == 2:
        try:
            return int(birth[:4])   # All '%Y/%m/%d' format
        except:
            return int(birth[-4:])  # All '%d/%m/%Y' format
    return np.nan

# See https://www.guidetofootball.com/tactics/playing-positions/ for more info
def clean_position(d):
    if 'GK' == d:    # GK
        return 'GK'
    elif 'B' in d:   # CB, LB, RB
        return 'DF'
    elif 'M' in d:   # CM, LM, LCM, LDM
        return 'MF'
    else:            # CS, LS, RS
        return 'FW'
```

```python
def performace_f(x, low_perf, high_perf):
    if x >= high_perf:
        return 'high performance'
    elif x < high_perf and x > low_perf:
        return 'middle performance'
    else:
        return 'low performance'
```

```python
low_perf = np.percentile(players['overall'].dropna(), 25)
high_perf = np.percentile(players['overall'].dropna(), 75)
players['performance'] = players['overall'].dropna().apply(performace_f, args=(low_perf, high_perf))
```

# Data Analysis

## Overall Trend Since 1950s

In order to discover players' heights trend, we made regression models for every different position of players. The result below shows that as players' heights increase as year goes by. Also, the p-value for all the regressions are less than 0.05, whcih implies that height is statistically significant factor in predicting players' overall performance. By looking at the plots with data height vs. year and linear regression line, and summary of regression results, we can conclude that players' heights and year has positive relationship for every position of soccer players.

```python
# Year x: Height: y

table = pd.DataFrame(columns=["GK", "DF", "MF", "FW"], index=["p_value", "coefficient"])
for pos in ["GK", "DF", "MF", "FW"]:
    positions = player_history[player_history.position == pos]
    mod = sm.OLS(positions.height, positions.year)
    res = mod.fit()
    pred_ols = res.get_prediction()
    display(res.summary())
    sns.lineplot(data=positions, x='year', y='height', label=pos)
    sns.lineplot(x=positions.year, y=res.predict())
```

## Correlation between height and preformance

In order to discover relationship between players' heights and their performance (overall score), we made regression models for every different position of players. The result below shows that as players' heights increase, their overall performance increases. Also, the p-value for all the regressions are less than 0.05, whcih implies that height is statistically significant factor in predicting players' overall performance. By looking at the plots with data points regarding height vs. overall and linear regression line, and summary of regression results, we can conclude that players' heights and their overall performance has positive relationship for every position of soccer players.

```python
fig, axes = plt.subplots(2, 2, constrained_layout=True)
fig.set_size_inches(20, 5)
for i, pos in enumerate(["FW", "MF", "DF", "GK"]):
    positions = players[players.position == pos]
    mod = sm.OLS(positions.overall, positions.height)
    res = mod.fit()
    display(res.summary())

    data = positions
    axes[i // 2, i % 2].scatter(positions.height, positions.overall)
    axes[i // 2, i % 2].plot(positions.height, res.predict())
    axes[i // 2, i % 2].set_title(pos)
    axes[i // 2, i % 2].set_xlabel("height")
    axes[i // 2, i % 2].set_ylabel("overallj")
```

## Correlation between BMI and performance

During the EDA part for weight, we could not really find out any relationship between weight and overall performance. Thus, we decided to calculate BMI which includes both height and weight to see if we could find the relationship between BMI and overall performance. Same as the previous model, we made linear regression models for every different positions of players. The results indicate that every models we made have positive coefficient for BMI, and the p-value was less than 0.05, which means that BMI is a statistically significant factor for overall performance. To conclude, we found a positive relationship between  bmi and  overall performance of players in every positions.

```python
players['BMI'] = round(players['weight'] / (players['height'])**2, 4)
fig, ax =plt.subplots(2,2, constrained_layout=True)
fig.set_size_inches(20, 5)

for i, pos in enumerate(["FW", "MF", "DF", "GK"]):
    positions = players[players['position'] == pos]
    mod = sm.OLS(positions.overall, positions.BMI)
    res = mod.fit()
    display(res.summary())
    sns.lineplot(x=positions.BMI, y=res.predict(), ax=ax[i // 2, i % 2])
    sns.scatterplot(data=positions, x="BMI", y="overall", ax=ax[i // 2, i %2]).set(title=pos)
```

# Limitation

The biggest limitation of our study comes from the ambiguous nature of the qualities we tried to quantify: "Overall Performance" of players that does not perfectly reflect the performance of players in the real nature of the league. We could only account for the numerical value that is provided from the FIFA Gaming series that we are uncertain of their method of calculation of players’ performance.

In addition, the presence of a limited pool of data or missing biological features of weight from 1950-2016 restricts our scope of BMI analysis since it has limitations with insufficient data of players that barred our research from deeper analysis with a wider range of players who played earlier than 2016.

Also, more limitations emerged from the presence of confounders affecting the independent (player’s height) and dependent (overall performance) variables in our analysis. Many outside factors impact on our independent variable (players’ height) besides years they played in accordance to their region of birth, ethnicity, etc. Moreover, our dependent variable (player’s performance) may also be influenced by many other factors besides just height and weight because there are many unaddressed factors such as agility, shooting, pace, strength, and passing that may affect players’ performance. Thus, our independent variables may just be one of many factors, and their impact is too small to make statistical significance.
